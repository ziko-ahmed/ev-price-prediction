```python
df['FastCharge_KmH'] = df['FastCharge_KmH'].replace('-', np.NAN)

df['FastCharge_KmH'] = pd.to_numeric(df['FastCharge_KmH'], errors='coerce')

df['FastCharge_KmH'] = df['FastCharge_KmH'].astype('Int64')
```

```python
num_cols = ['AccelSec','TopSpeed_KmH','Range_Km','Efficiency_WhKm','FastCharge_KmH','Seats','PriceEuro']
corr_matrix = df[num_cols].corr()
sns.heatmap(corr_matrix, linewidths=1, linecolor='white', annot=True)
```
